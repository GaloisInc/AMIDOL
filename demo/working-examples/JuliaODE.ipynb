```julia
using DifferentialEquations
using Plots

function sir(du,u,p,t)
  s,i,r = u
  β, γ = p
  du[1] = ds = -β*s*i/(s+i+r)
  du[2] = di = -γ*i + β*s*i/(s+i+r)
  du[3] = dr = γ*i
end

u0 = [500.0; 10.0; 0.0]
tspan = (0.0,100.0)
p = [2.0 * 1.0/3.0, 1.0/3.0]
prob = ODEProblem(sir, u0, tspan, p)
sol = solve(prob)

plot(sol)
```
