```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np
from sklearn.metrics import mean_squared_error
%matplotlib inline

plt.style.use('ggplot')
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),
    (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),
    (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),
    (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),
    (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]

for i in range(len(tableau20)):
    r, g, b = tableau20[i]
    tableau20[i] = (r / 255., g / 255., b / 255.)


def weekTransform(row):
    if (row["WEEK"] >= 40):
        week = row["WEEK"] - 40
    else:
        week = row["WEEK"] + 12
    return(week)

df20072008 = pd.read_csv('WHO_NREVSS_Combined_2007_2008.csv', header=1)
df20102011 = pd.read_csv('WHO_NREVSS_Combined_2010_2011.csv', header=1)

df20072008["idx: t"] = df20072008.apply(weekTransform, axis=1)
df20102011["idx: t"] = df20102011.apply(weekTransform, axis=1)
```

```python
Y = {}
X = {}

Y['B', 1] = df20072008['B']
X['B', 1] = df20072008['idx: t']

Y['B', 2] = df20102011['B']
X['B', 2] = df20102011['idx: t']

Y['A (H1)', 1] = df20072008['A (H1)']
X['A (H1)', 1] = df20072008['idx: t']

Y['A (H1)', 2] = df20102011['A (H1)']
X['A (H1)', 2] = df20102011['idx: t']

Y['A (H3)', 1] = df20072008['A (H3)']
X['A (H3)', 1] = df20072008['idx: t']

Y['A (H3)', 2] = df20102011['A (H3)']
X['A (H3)', 2] = df20102011['idx: t']


plt.figure(figsize=(12,6))
plt.plot(X['B',1], Y['B',1], ls="--", lw=2.5, color=tableau20[0], alpha=0.3)
plt.scatter(X['B',1], Y['B',1], color=tableau20[0])
plt.plot(X['B', 2], Y['B', 2], lw=2.5, color=tableau20[1], alpha=0.3)
plt.scatter(X['B',2], Y['B',2], color=tableau20[1])
######
plt.plot(X['A (H1)',1], Y['A (H1)',1], ls="--", lw=2.5, color=tableau20[2], alpha=0.3)
plt.scatter(X['A (H1)',1], Y['A (H1)',1], color=tableau20[2])
plt.plot(X['A (H1)', 2], Y['A (H1)', 2], lw=2.5, color=tableau20[3], alpha=0.3)
plt.scatter(X['A (H1)',2], Y['A (H1)',2], color=tableau20[3])
######
plt.plot(X['A (H3)',1], Y['A (H3)',1], ls="--", lw=2.5, color=tableau20[4], alpha=0.3)
plt.scatter(X['A (H3)',1], Y['A (H3)',1], color=tableau20[4])
plt.plot(X['A (H3)', 2], Y['A (H3)', 2], lw=2.5, color=tableau20[5], alpha=0.3)
plt.scatter(X['A (H3)',2], Y['A (H3)',2], color=tableau20[5])
plt.title("CDC Infection Data", fontsize=12)
plt.xlabel("Week of Flu Season", fontsize=10)
plt.ylabel("Infected Incidence", fontsize=10)
```
