```python
from bs4 import BeautifulSoup
import requests

URL = "https://coronavirus.1point3acres.com/en"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') 

mydivs = soup.findAll("div", {"class": "state-table"})
rows = mydivs[0].findAll("div", {"class": "row"})

columns = ['State', 'Confirmed', 'Deaths', 'Mortality']

for row in rows:
    spans = row.findAll("span")
    columns = []
    for span in spans:
        columns.append(span.contents[0])
    print(columns)
```
