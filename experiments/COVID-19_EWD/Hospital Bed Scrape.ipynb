```python
import pandas as pd

URL = "American Hospital Directory - Table of Search Results.htm"

content = ''

with open(URL, "r+") as fin:
    content = fin.read()
  
df = pd.read_html(content)[5].drop(['Reports'], axis=1)

df.to_csv('hospitals_beds.csv')
```

```python
import geocoder

def getCounty(row):
    location = row['City'] + ", " + row['State']
    results = geocoder.google(location)
    print(results)

df.apply(getCounty, axis=1)
```
